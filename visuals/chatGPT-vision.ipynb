```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_chart():
    # Data
    data = {
        'H2 Emissions Limit': [0.45, 1.5, 3.38, 4.5],
        'Region': ['USA', 'USA', 'EU', 'USA'],
        'Median': [45, 25, 20, 55],
        'Upper Quartile': [52, 35, 30, 70],
        'Lower Quartile': [38, 15, 10, 40]
    }
    df = pd.DataFrame(data)

    # Plotting
    fig, ax = plt.subplots(figsize=(10, 8))
    
    colors = {'USA': 'green', 'EU': 'blue'}
    for idx, row in df.iterrows():
        ax.boxplot([range(int(row['Lower Quartile']), int(row['Upper Quartile']+1))],
                   positions=[row['H2 Emissions Limit']],
                   widths=0.5,
                   patch_artist=True,
                   boxprops=dict(facecolor=colors[row['Region']]),
                   medianprops=dict(color='black'),
                   vert=True)
    
    # Enhancing the Background Colors
    ax.axhspan(0, 11, facecolor='blue', alpha=0.2)
    ax.axhspan(11, 48, facecolor='yellow', alpha=0.2)
    
    # Annotations and lines for solar utility and onshore wind
    ax.axhline(y=48, color='black', linestyle='--')
    ax.text(4.7, 50, '48 g CO2-eq/kWh\n(solar utility)', va='center')
    
    ax.axhline(y=11, color='black', linestyle='--')
    ax.text(4.7, 13, '11 g CO2-eq/kWh\n(onshore wind)', va='center')

    # Labels, title, and grid
    ax.set_xlabel('Hydrogen Emissions Limit [kg CO2-eq/kg H2]')
    ax.set_ylabel('Electricity Carbon Intensity [g CO2-eq/kWh]')
    ax.set_title('GWP = 11.5 (~GWP100)\nelectrolysis efficiency = (50 kWh el/1 kg H2)\n{0,2.5,5,7.5,10}% H2 emissions')
    ax.set_xlim(0, 5)
    ax.set_xticks(df['H2 Emissions Limit'])
    ax.set_xticklabels(['0.45 (USA)', '1.5 (USA)', '3.38 (EU)', '4.5 (USA)'])
    ax.grid(axis='y', linestyle='--')

    # Display
    plt.show()

# Call the function
plot_chart()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def plot_circle_with_wedge():
    fig, ax = plt.subplots()

    # Plot a circle
    circle = patches.Circle((0.5, 0.5), 0.4, color='grey', fill=False, label='Circle')
    ax.add_patch(circle)

    # Plot a wedge inside the circle
    theta1, theta2 = 0, 60  # angles in degrees
    wedge = patches.Wedge(center=(0.5, 0.5), r=0.4, theta1=theta1, theta2=theta2, color='grey', label='Wedge')
    ax.add_patch(wedge)

    # Setting aspect ratio and limits
    ax.set_aspect('equal', 'box')
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.axis('off')  # Turn off the axis

    # Add title and legend
    ax.set_title('Circle with a Wedge')
    ax.legend(loc='upper left', bbox_to_anchor=(1, 1))

    plt.tight_layout()
    plt.show()

plot_circle_with_wedge()
```
