```python
import math
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
def plot_sin():
    x = np.linspace(-np.pi, np.pi, 1000)
    y = np.sin(x)
    plt.plot(x, y)
    plt.grid(True)
    plt.show()
```

```python
plot_sin()
```
